# Exploratory Data analysis of Bangalore city outlets

# Importing Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import re
```

# Loading data

```python
df=pd.read_csv('Bangalore Outlet Details.csv')
df.head()
```

# Get to know about data

```python
df.shape
```

```python
df.columns
```

```python
df.dtypes
```

# Cleaning
## Ratings and Cost for two- Convert to numeric type

```python
df.loc[df.Rating=='--','Rating']=np.nan
```

```python
df.Rating=df.Rating.astype(float)
```

```python
df.Cost_for_Two.str.strip('₹ ').astype(int)
```

```python
df.Cost_for_Two=df.Cost_for_Two.apply(lambda x:re.search(r'\d+',x).group()).astype(int)
```

```python
df.dtypes
```
